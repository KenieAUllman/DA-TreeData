```python
import ipywidgets as widgets 
from ipywidgets import Output 
from IPython.display import display 

out = Output()
```

```python
data_tree.columns = data_tree.columns.str.replace(" ", "_")
```

```python
 cleaned_data = data_tree.dropna(axis=0)
```

```python
zip_unique = list(cleaned_data.Zip_Codes.unique())
zip_unique
```

```python
dd_zips = widgets.Dropdown(options=zip_unique)
dd_zips
display(dd_zips)
```

```python
def dd_zip_observer(bunch):
    out.clear_output()
    with out:
        display(cleaned_data[zip_unique == bunch.new])
```

```python
dd_zips.observe(dd_zip_observer, names="value")
```
